```python
import pandas as pd

# --- EDIT THESE:
input_file = "all_timelines_merged.csv"
output_file = "transformer_preprocessed.csv"
```

```python
mapping_adaptive = {
    'adaptive-state_A_category': {
        "(1) Calm/ laid back": 1,
        "(3) Sad, Emotional pain, grieving": 2,
        "(5) Content, happy, joy, hopeful": 3,
        "(7) Vigor / energetic": 4,
        "(9) Justifiable anger/ assertive anger, justifiable outrage": 5,
        "(11) Proud": 6,
        "(13) Feel loved, belong": 7,
    },
    'adaptive-state_B-O_category': {
        "(1) Relating behavior": 1,
        "(3) Autonomous or adaptive control behavior": 2,
    },
    'adaptive-state_B-S_category': {
        "(1) Self care and improvement": 1,
    },
    'adaptive-state_C-O_category': {
        "(1) Perception of the other as related": 1,
        "(3) Perception of the other as facilitating autonomy needs": 2,
    },
    'adaptive-state_C-S_category': {
        "(1) Self-acceptance and compassion": 1,
    },
    'adaptive-state_D_category': {
        "(1) Relatedness": 1,
        "(3) Autonomy and adaptive control": 2,
        "(5) Competence, self esteem, self-care": 3,
    }
}

mapping_maladaptive = {
    'maladaptive-state_A_category': {
        "(2) Anxious/ fearful/ tense": 1,
        "(4) Depressed, despair, hopeless": 2,
        "(6) Mania": 3,
        "(8) Apathetic, don’t care, blunted": 4,
        "(10) Angry (aggression), disgust, contempt": 5,
        "(12) Ashamed, guilty": 6,
        "(14) Feel lonely": 7,
    },
    'maladaptive-state_B-O_category': {
        "(2) Fight or flight behavior": 1,
        "(4) Over controlled or controlling behavior": 2,
    },
    'maladaptive-state_B-S_category': {
        "(2) Self harm, neglect and avoidance": 1,
    },
    'maladaptive-state_C-O_category': {
        "(2) Perception of the other as detached or over attached": 1,
        "(4) Perception of the other as blocking autonomy needs": 2,
    },
    'maladaptive-state_C-S_category': {
        "(2) Self criticism": 1,
    },
    'maladaptive-state_D_category': {
        "(2) Expectation that relatedness needs will not be met": 1,
        "(4) Expectation that autonomy needs will not be met": 2,
        "(6) Expectation that competence needs will not be met": 3,
    }
}
```

```python
df = pd.read_csv(input_file)

# Map the columns
def process_col(col, mapping, new_suffix):
    def map_value(x):
        if pd.isnull(x) or str(x).strip() == "":
            return 0
        return mapping.get(str(x).strip(), 0)
    if col.startswith("adaptive-state_"):
        base = col.replace("adaptive-state_", "").replace("_category", "")
    else:
        base = col.replace("maladaptive-state_", "").replace("_category", "")
    new_col = f"{base}_{new_suffix}"
    df[new_col] = df[col].apply(map_value)
    return new_col

new_columns = []

for col, mapping in mapping_adaptive.items():
    if col in df.columns:
        new_col = process_col(col, mapping, "adaptive")
        new_columns.append(new_col)

for col, mapping in mapping_maladaptive.items():
    if col in df.columns:
        new_col = process_col(col, mapping, "maladaptive")
        new_columns.append(new_col)

cols_to_keep = ["post_text"] + new_columns
print(new_columns)
df_final = df[cols_to_keep]

df_final.to_csv(output_file, index=False)
print(f"Done! Output written to {output_file}")
```
